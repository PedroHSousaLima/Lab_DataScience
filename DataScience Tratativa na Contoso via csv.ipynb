### Realizando a importação dos arquivos

```python
import pandas as pd

contoso_prd_df = pd.read_csv(r'Contoso - Cadastro Produtos.csv', sep=';', encoding='utf-8')
contoso_vnd17_df = pd.read_csv(r'Contoso - Vendas - 2017.csv', sep=';', encoding='utf-8')
contoso_loj_df = pd.read_csv(r'Contoso - Lojas.csv', sep=';', encoding='utf-8')
contoso_client_df = pd.read_csv(r'Contoso - Clientes.csv', sep=';', encoding='utf-8')
```

### Mesclar dados

```python
# Traduzindo as ID's de Vendas
contoso_client_df = contoso_client_df[['ID Cliente', 'Primeiro Nome', 'Sobrenome', 'E-mail']]
contoso_prd_df = contoso_prd_df[['ID Produto', 'Categoria']]
contoso_loj_df = contoso_loj_df[['ID Loja', 'Nome da Loja']]
#display(contoso_prd_df)

contoso_vnd17_df = contoso_vnd17_df.merge(contoso_prd_df, on = 'ID Produto')
contoso_vnd17_df = contoso_vnd17_df.merge(contoso_loj_df, on = 'ID Loja')
contoso_vnd17_df = contoso_vnd17_df.merge(contoso_client_df, on = 'ID Cliente')

#display(contoso_vnd17_df)
contoso_vnd17_df.info()
```

### Renomear coluna

```python
contoso_vnd17_df = contoso_vnd17_df.rename(columns={'Categoria': 'Produto'})
display(contoso_vnd17_df)
```

### Transformação das colunas de Data

```python
contoso_vnd17_df['Data da Venda'] = pd.to_datetime(contoso_vnd17_df['Data da Venda'], format='%d/%m/%Y')
contoso_vnd17_df['Ano da Venda'] = contoso_vnd17_df['Data da Venda'].dt.year
contoso_vnd17_df['Mes da Venda'] = contoso_vnd17_df['Data da Venda'].dt.month
contoso_vnd17_df['Dia da Venda'] = contoso_vnd17_df['Data da Venda'].dt.day
contoso_vnd17_df['Data do Envio'] = pd.to_datetime(contoso_vnd17_df['Data do Envio'], format='%d/%m/%Y')
contoso_vnd17_df.info()
```

### Concatenar colunas

```python
contoso_vnd17_df['Full_Name'] = contoso_vnd17_df['Primeiro Nome'] +' '+ contoso_vnd17_df['Sobrenome']
display(contoso_vnd17_df)
```

### Top Ranking de Clientes

```python
frenquencia_clientes = contoso_vnd17_df['Full_Name'].value_counts()
frenquencia_clientes[:5].plot(figsize=(15,5))
```

### Top Ranking das Vendas por Loja

```python
vnd17_loj = contoso_vnd17_df.groupby('Nome da Loja').sum()
vnd17_loj = vnd17_loj[['Quantidade Vendida']]

maior_valor = vnd17_loj['Quantidade Vendida'].max()
melhor_loja = vnd17_loj['Quantidade Vendida'].idxmax()
vnd17_loj = vnd17_loj.sort_values('Quantidade Vendida', ascending=False)
vnd17_loj[:5].plot(figsize=(15,5), kind='bar')
print(f'A campeã de vendas foi {melhor_loja} com um total de {maior_valor}')

display(vnd17_loj)

```

### Produto que menos vendeu

```python
produto = contoso_vnd17_df.groupby('Produto').sum()
produto = produto[['Quantidade Vendida']]

menor_valor = produto['Quantidade Vendida'].min()
min_produto = produto['Quantidade Vendida'].idxmin()
produto = produto.sort_values('Quantidade Vendida', ascending=True)
print(f'O Produto menos vendido foi {min_produto} com um total de {menor_valor} unidades')
produto[:5].plot(figsize=(15,5), kind='bar')

display(produto)

```
